```python
import string

class TextContainer():
    """A simple book model consisting of chapters, which in 
    turn consist of paragraphs."""

    def __init__(self, text):
        """Initialize title, the author, and the chapters."""
        self.text = text
    
    def count_words(self):
        wordlist = self.text.split()      
        return len(wordlist)    
    
    def number_of_chars(self):
        letter_list = list(self.text)
        count = 0
        for letter in letter_list:
            if letter.isalpha():
                #print(letter)
                count = count + 1
        return count
    
    def number_of_letter(self):
        count = 0
        for letter in self.text:
            if letter in string.ascii_letters:
                #print(letter)
                count = count + 1
        return count
    
    def number_of_punctuation(self):
        count = 0
        for letter in self.text:
            if letter in string.punctuation:
                #print(letter)
                count = count + 1
        return count
        
            
```

```python
string.ascii_letters
```

```python
string.punctuation
```

```python
tc = TextContainer("Dette er en tekst!?")
```

```python
# Number of words
tc.count_words()
```

```python
# Number of chars
tc.number_of_chars()
```

```python
tc.number_of_letter()
```

```python
tc.number_of_punctuation()
```
