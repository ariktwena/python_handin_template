```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.firefox.options import Options
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

import bs4
import json
```

```python
url = 'https://www.dba.dk'
```

```python
def dba_interaction(url, searchPhrase):
    profile = webdriver.FirefoxProfile()
    profile.set_preference("general.useragent.override", "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:81.0) Gecko/20100101 Firefox/81.0")
    
    # headless is needed here because we do not have a GUI version of firefox
    options = Options()
    options.headless = True
    # driver = webdriver.Firefox(options=options, executable_path=r'/tmp/geckodriver')
    browser = webdriver.Firefox(options=options)
    
    # browser = webdriver.Firefox()
    browser.get(url)
    browser.implicitly_wait(2)
    button = browser.find_element_by_id('onetrust-reject-all-handler')
    button.click()
    
    #Search field
    search_field = browser.find_element_by_id('searchField')
    print(search_field)
    search_field.send_keys(searchPhrase)
    search_field.submit()
    
    sleep(3)
    
    #Click kbh
    button = browser.find_element_by_xpath('//*[@id="regionNavigatorDiv"]/ul[1]/li[1]/span/a')
    button.click()

    sleep(3)
    
    button = browser.find_element_by_xpath('//*[@id="srpNavigators"]/div[4]/h4')
    button.click()

    sleep(3)
    
    button = browser.find_element_by_xpath('//*[@id="srpNavigators"]/div[4]/div/ul/li[1]/span')
    button.click()

    sleep(3)
    
    return  browser.page_source
```

```python
page_source = dba_interaction(url, 'macbook pro')
```

```python
page_source
```

```python
def find_elements(page):
    soup = bs4.BeautifulSoup(page, 'html.parser')
    #print(soup)
    elements = soup.select('tr[class="dbaListing listing"] > td[class="mainContent"] > script')
    #event_cells = soup.select(selector)
    #return event_cells
    return elements
```

```python
event_cells = find_elements(page_source)
```

```python
event_cells
```

```python
json_list = []
for event_cell in event_cells:
    elemets = str(event_cell).split('\n')
    json_string_elements = elemets[1:-1]
    json_string = ''.join(json_string_elements)
    the_json = json.loads(json_string)
    json_list.append(the_json)
    
```

```python
json_list
```

```python
for json_string in json_list:
    print(json_string["name"] + " price: " + json_string["offers"]["price"] + " kr")
```

```python
tuple_list = []
for json_string in json_list:
    tuple_list.append((json_string['name'], json_string['image'], json_string['url'], json_string["offers"]["price"]))
    
```

```python
def convert_tuple(single_tuple):
    
    string = ''
    
    product = single_tuple[0]
    image = single_tuple[1]
    url = single_tuple[2]
    price = single_tuple[3]
    
    string = f'<tr><td>{product}</td><td>{image}</td><td>{url}</td><td>{price}</td></tr>'
    
    print (string)
    
    return string 
```

```python
the_tuple_list = convert_tuple(tuple_list_sorted)
```

```python
the_tuple_list
```

```python
def create_table(tuple_list, function):
    table_data = '<table><tr><th>Product</th><th>image</th><th>url</th><th>price</th></tr>'
    
    map_data = map(function, tuple_list)
    string_from_map_data = ''.join(map_data)
    
    table_data = table_data + string_from_map_data + '</table>'
    
    return table_data
```

```python
table_data = create_table(tuple_list_sorted, convert_tuple)
```

```python
def multithreading(func, args, workers=10):
    
    with ThreadPoolExecutor(workers) as ex:
        res = ex.map(func, args)
        print(res)
    return list(res)
```

```python
def get_number_by_url(url):
    profile = webdriver.FirefoxProfile()
    profile.set_preference("general.useragent.override", "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:81.0) Gecko/20100101 Firefox/81.0")
    
    # headless is needed here because we do not have a GUI version of firefox
    options = Options()
    options.headless = True
    # driver = webdriver.Firefox(options=options, executable_path=r'/tmp/geckodriver')
    browser = webdriver.Firefox(options=options)
    
    # browser = webdriver.Firefox()
    browser.get(url)
    #browser.implicitly_wait(2)
   
    
    #Click phone
    button = browser.find_element_by_xpath('//*[@id="business-card"]/div[1]/div[2]/div[2]/div/button/span[1]/span')
    button.click()

    sleep(3)
    
    
    soup = bs4.BeautifulSoup(browser.page_source, 'html.parser')
    
    element = soup.select('ul[class="dba-MuiList-root dba-MuiList-padding"] > li[class="dba-MuiListItem-root dba-MuiMenuItem-root dba-dba-jss-3 dba-MuiMenuItem-gutters dba-MuiListItem-gutters"] > button[class="dba-MuiButtonBase-root dba-MuiButton-root dba-MuiButton-outlined ContactButton_contact-button__number__ARGM8 dba-MuiButton-outlinedPrimary"]')
    
    phone_num_reg_ex_pattern_obj = re.compile(r'\d{8}')
    
    match_obj = phone_num_reg_ex_pattern_obj.search(str(element[0]))
    print(match_obj)
    print(match_obj.group())
    phone_number = match_obj.group()
    
    return phone_number
```

```python
phone_number = get_number_by_url('https://www.dba.dk/macbook-pro-apple-macbook-pro/id-121387288/')
```
